```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
sdc = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
sdc.head()
```

```python
#Calculate the average math score
avg_math_score_f = (("{:,.2f}%").format(sdc["math_score"].mean()))
#unformatted version so it doesn't break the % passing code
avg_math_score = sdc["math_score"].mean()
avg_math_score_f
```

```python
#Calculate the average reading score
avg_reading_score_f = (("{:,.2f}%").format(sdc["reading_score"].mean()))
#unformatted version so it doesn't break the % passing code
avg_reading_score = sdc["reading_score"].mean()
print(avg_reading_score_f)
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
passing_rate_f = (("{:.2f}%").format((avg_math_score+avg_reading_score)/2))
passing_rate = (avg_math_score+avg_reading_score)/2
passing_rate_f
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_score = sdc["math_score"]
counter = 0
for row in math_score:
    if row >=70:
        counter +=1
passing_math_count = sdc[(sdc["math_score"] >= 70)].count()["student_name"]
passing_math_perc = passing_math_count / float(total_students) * 100
passing_math_percentage = ("{:.2f}%".format(passing_math_perc))
passing_math_percentage
```

```python
##**change to same format as cell above once fixed
# #Calculate the percentage of students with a passing reading score (70 or greater)
read_score = sdc["reading_score"]
counter2 = 0
for row in read_score:
    if row >=70:
        counter2 +=1
passing_read_count = sdc[(sdc["reading_score"]>=70)].count()["student_name"]
passing_read_perc = passing_read_count/float(total_students)*100
passing_read_percentage = ("{:.2f}%".format(passing_read_perc))
passing_read_percentage
```

```python
#Create a dataframe to hold the above results
    #Optional: give the displayed data cleaner formatting per_student_budget = float(total_budget/total_students)
district_summary = pd.DataFrame({
    'Total Schools': [total_schools],
    'Total Students': [total_students_f],
    'Total Budget': [total_budget_f],
    'Avg. Math Score':[avg_math_score_f],
    'Avg. Reading Score':[avg_reading_score_f],
    'Overall Passing Rate':[passing_rate_f],
    'Math Pass Percent':[passing_math_percentage],
    'Reading Pass Percent':[passing_read_percentage]
})
district_summary.head()
```

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_types = school_data.set_index(["school_name"])["type"]
school_students = sdc['school_name'].value_counts()
school_budget = school_data.groupby(['school_name']).budget.min()
per_student_budget = school_budget/school_students
school_avg_math = sdc.groupby(['school_name']).mean()['math_score']
school_avg_read = sdc.groupby(['school_name']).mean()['reading_score']
group_keys = (sdc.school_name)
school_math_p = ((sdc['math_score']>=70).groupby(group_keys).sum()/school_students)*100
school_read_p = ((sdc['reading_score']>=70).groupby(group_keys).sum()/school_students)*100
school_overall_pr = ((school_math_p + school_read_p)/2)
```

```python
##DF #2
school_summary = pd.DataFrame({
    'School Type': school_types,
    'Total Students': school_students,
    'Total School Budget':school_budget,
    'Per Student Budget':per_student_budget,
    'Average Math Score':school_avg_math,
    'Average Reading Score':school_avg_read,
    '% Passing Math':school_math_p,
    '% Passing Reading':school_read_p,
    'Overall Passing Rate':school_overall_pr
})
school_summary.head()
```
